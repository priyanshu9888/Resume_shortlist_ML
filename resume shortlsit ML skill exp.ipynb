```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB


require_skills = ["Python","flask","NLTK","ML"]
require_certification = ["Certified Data Analyst","Certified java Scientist","Certified Data Scientist"]
require_qualification = ["Bachelor's degree in Computer Science","Master's degree in Computer Science"]
require_experience = 5
import json

applications = {
    "total_exp": [2, 4, 5, 3, 6, 3,10,1],
    "skills": [
        ["Python", "Java","NLTK","flask"],
        ["R", "Data analysis"],
        ["Java", "Spring Framework", "Hibernate"],
        ["Python", "Data analysis", "Data visualization"],
        ["SQL", "Data modeling", "Oracle"],
        ["Python", "Data modeling", "NLTK","NLP"],
        ["Java","spring","HTML","CSS"],
        ["Python","ML"]
    ],

        "exp_skill_name": ["java","python", "go", "sql", "R","NLTK","Java","Python"],
        "skill_exp": [2, 4, 8, 5, 3,4,2,3],

    "certifications": [
        [],
        ["Certified Data Analyst"],
        ["Certified Spring Framework Developer"],
        ["Certified Data Visualization Expert"],
        ["Certified Oracle Database Developer"],
        [],
        ["Certified Java Developer"],
        ["Certified Python Developer"]
        
    ],
    "qualification": [
        "Bachelor's degree in Computer Science",
        "Master's degree in Statistics",
        "Bachelor's degree in Computer Engineering",
        "Master's degree in Data Science",
        "Bachelor's degree in Computer Science",
        "Bachelor's degree in Computer Science",
        "Master's degree in Computer Science",
        "Master's degree in Computer Science"
        
    ]
}

df = pd.DataFrame(applications)




skill_range = {
    (0, 1): 0,
    (2, 4): 1,
    (4, 8): 1,
}

def skill_score(skill, skill_experience):
    for r in skill_range:
        if skill_experience >= r[0] and skill_experience < r[1]:
            return skill_range[r]
    return 0

exp_skill_name = df["exp_skill_name"]
skill_exp = df["skill_exp"]
# skill_scores = {}

for i in range(len(exp_skill_name)):
    skill_scores[exp_skill_name[i]] = skill_score(exp_skill_name[i], skill_exp[i])

print(skill_scores)



def count_matched_skills(skills):
    return len(set(require_skills).intersection(set(skills)))
# def count_matched_skills(skills):
#     matched = len(set(require_skills).intersection(set(skills)))
#     return f"{matched}/{len(require_skills)}"


# def get_skills_ratio(skills):
#     return count_matched_skills(skills) / len(require_skills)


def certification_match(certifications):
    if any(cert in require_certification for cert in certifications):
        return 1
    else:
        return 0


def qualification_match(qualification):
    if qualification in require_qualification:
        return 1
    else:
        return 0
    
def experience_match(exp):
    if exp <= require_experience:
        return 1
    else:
        return 0
def get_skill_score(skill):
    return skill_scores.get(skill, 0)
     
df["skillsMatched"] = df["skills"].apply(count_matched_skills)
df["skill_scores"] = df["exp_skill_name"].apply(get_skill_score)
df["certificationMatch"] = df["certifications"].apply(certification_match)

df["qualificationMatch"] = df["qualification"].apply(qualification_match)
df["experienceMatch"] = df["total_exp"].apply(experience_match)
df.columns
# df["status"]=y
X = df[["skill_scores","skillsMatched", "certificationMatch", "qualificationMatch", "experienceMatch"]]

X
df

y = pd.Series([1,1,0,1,0,1,0,0])
X
model = MultinomialNB()
model.fit(X, y)
model
```
